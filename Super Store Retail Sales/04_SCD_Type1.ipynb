```python
# Read the Category Dimension

dim_category = spark.table(
    "`Superstore Data Engineering`.SuperstoreLakehouse.dbo.dim_category"
)

display(dim_category)
```

```python
updated_dim = (
    dim_category.replace(
        "Technology",
        "Technology & Electronics",
        subset=["category"]
    )
)

display(updated_dim)
```

```python
updated_dim.write \
.mode("overwrite") \
.format("delta") \
.saveAsTable("dim_category")
```
